```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,3,4,5,6,7])
y = np.array([20,37,56,75,99,76])
colors = ["r","b","k","y","c","purple"]
sizes = np.array([20,300,100,487,550,60])
plt.scatter(x,y,s = sizes,c = colors)
plt.show()
```
